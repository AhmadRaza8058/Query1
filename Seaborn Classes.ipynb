```python
dj['Date']= pd.to_datetime(dj['Date'])  # df = pd.to_datetime(dj['Date'])      DOUBT
```

```python
fig, ax = plt.subplots(figsize=(21, 6))
fig = sns.lineplot(x='Date', y='Price', data=dj)
x_dates = dj['Date'].dt.strftime("%Y-%m-%d").sort_values().unique()
ax.set_xticklabels(labels=x_dates, rotation=20, ha='right')
plt.show()
```
